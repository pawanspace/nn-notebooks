#### Simple notebook to implement and plot functions

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import torch.nn.functional as F
import torch as t
%matplotlib inline
```

```python
def inflection_point(x):
    return -4 * np.sin(2 * x)

numbers = np.linspace(-1 * np.pi/4, 5* np.pi/4, 100)
numbers
def generate_inflection_point_each_iteration(numbers):
    for number in numbers:
        yield inflection_point(number)

inflection_points = generate_inflection_point_each_iteration(numbers)

def plot(inflection_points):
    plt.plot(inflection_points)
    plt.show()
plot(np.fromiter(inflection_points,  dtype=float))
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

numbers = np.linspace(-10, 10, 100)

def generate_sigmoid_output(numbers):
    for number in numbers:
        yield sigmoid(number)

def generate_tanh_output(numbers):
    for number in numbers:
        yield np.tanh(number)

def generate_softmax_output(numbers):
    for number in numbers:
        yield np.softmax(number)

def generate_sigmoid_derivative_output(numbers):
    for number in numbers:
        yield sigmoid_derivative(number)

#sigmoid
s_output = np.fromiter(generate_sigmoid_output(numbers), dtype=float)
plt.style.use('_mpl-gallery')
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots()
ax.plot(numbers, s_output, linewidth=1.0)

#tanh
s_output = np.fromiter(generate_tanh_output(numbers), dtype=float)
plt.style.use('_mpl-gallery')
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots()
ax.plot(numbers, s_output, linewidth=1.0)

#softmax
s_output = F.softmax(t.tensor(numbers), dim=0, dtype=float)
plt.style.use('_mpl-gallery')
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots()
ax.plot(numbers, s_output, linewidth=1.0)
```
